```python
train_len = len(train)
#combine two dataframes
df = pd.concat([train,test],axis=0)
df = df.reset_index(drop = True)
df.head()
```

```python
df.tail()
```

```python
##Find the NULL Values
df.isnull().sum()
```

```python
#Length of dataframe
len(df)
```

```python
##Drop or Delete the Column
df = df.drop(columns=['Cabin'],axis=1)
```

```python
#Fill the missing values using mean value of that Column = Numerical Column
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df['Age'].mean()
```

```python
df['Fare'].mean()
```

```python
df['Embarked'].mode()[0]
```

```python
#Fill the missing vaules using mode of that Column = Catergorical Column
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Embarked']
```

```python
df['Fare'] = np.log(df['Fare'] + 1)
```

```python
df['Fare']
```

```python
sns.distplot(df['Fare'])
```

```python
#Correlation Matrix
#corr = df.corr()

#plt.figure(figsize=(15,9))
#sns.heatmap(corr,annot=True,cmap='coolwarm')
df['Sex'] = df['Sex'].astype(str) 
df['Embarked'] = df['Embarked'].astype(str) 
```

```python
#Convert Sex and Embarked Column into interger
#That Convertion means Label Encoding
##Label Encoding
from sklearn.preprocessing import LabelEncoder
cols = ['Sex','Embarked']
le = LabelEncoder()

for col in cols:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
df = df.drop(columns=['Name','Ticket'],axis=1)
df.head()
```

```python
df.info()
```

```python
#Splitting data set 
print(train_len)
```

```python
train = df.iloc[:train_len, :]
test = df.iloc[train_len:, :]
```

```python
train.head()
```

```python
test.head()
```

```python
#input split
X = train.drop(columns=['PassengerId','Survived'],axis=1)
y = train['Survived']
```

```python
X.head()
```

```python
y.head()
```

```python
#Model training
from sklearn.model_selection import train_test_split,cross_val_score
def classify(model, x, y):
    x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=42)
    model.fit(x_train,y_train)
    print('Accuracy:',model.score(x_test, y_test))
    
    score = cross_val_score(model, x, y, cv=5)
    print('CV Score:',np.mean(score))
    
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model,X,y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model,X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model,X,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model,X,y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
classify(model,X,y)
```

```python
from lightgbm import LGBMClassifier
modellgbm = LGBMClassifier()
classify(modellgbm,X,y)
```

```python
from xgboost import XGBClassifier
modelxgb = XGBClassifier()
classify(modelxgb,X,y)
```

```python
X_test = test.drop(columns=['PassengerId','Survived'], axis=1)
```
